```python
three_keywords_df = pd.read_csv('data/three_keywords.csv')
#three_keywords_df.head()
three_keywords_df['month'] = pd.to_datetime(three_keywords_df['month'], format='%Y-%m') 
#print(three_keywords_df['month'])

# Find the month with the highest 'home_workout_worldwide'
peak_covid_month = three_keywords_df.loc[three_keywords_df['home_workout_worldwide'].idxmax(), 'month']
peak_covid_year = peak_covid_month.year 

# Find the month with the highest 'gym_workout_worldwide'
current_month = three_keywords_df.loc[three_keywords_df['gym_workout_worldwide'].idxmax(), 'month']
current_year = current_month.year

print(f"Peak COVID home workout interest: {peak_covid_year}")
print(f"Current peak gym workout interest: {current_year}") 
```
